```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib notebook
```

```python
data = pd.read_csv('Total Tub Energy Consumption by Building Edited.csv')
years = list(range(2000,2022))
#data.head()
```

```python
dunster_kitchen_MTCDE = data.iloc[[5]].values.tolist()[0]
dunster_kitchen_MTCDE = dunster_kitchen_MTCDE[4:47:2]

plt.figure()
plt.plot(years, dunster_kitchen_MTCDE)
plt.title('MTCDE in Dunster Kitchen Since 2000')
plt.xlabel('Year')
plt.ylabel('MTCDE')
```

```python
dunster_kitchen_kBtu = data.iloc[[5]].values.tolist()[0]
dunster_kitchen_kBtu = dunster_kitchen_kBtu[3:46:2]

plt.figure()
plt.plot(years, dunster_kitchen_kBtu)
plt.title('kBtu in Dunster Kitchen Since 2000')
plt.xlabel('Year')
plt.ylabel('kBtu')
```

```python
dunster_kitchen_ratio = list(map(lambda a,b: a / b, dunster_kitchen_kBtu, dunster_kitchen_MTCDE))
#dunster_kitchen_ratio = [a / b for a,b in zip(dunster_kitchen_kBtu, dunster_kitchen_MTCDE)]

plt.figure()
plt.plot(years, dunster_kitchen_ratio)
plt.title('kBtu / MTCDE in Dunster Kitchen Since 2000')
plt.xlabel('Year')
plt.ylabel('kBtu / MTCDE')
```
